## RNN Implementation

```python
import numpy as np
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
from keras import Sequential
```

```python
 # list
doc = [
    
        'I love this movie',
        'This is terrible',
        'Amazing experience',
        'Worst film ever',
        'Very good acting',
        'Not good at all'
    ]
labels = [1, 0, 1, 0, 1, 0]   # sentiment labels
```

```python
# A Tokenizer converts text (words) into numbers (tokens)
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer(oov_token='<nothing>')
```

```python
tokenizer.fit_on_texts(doc)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
sequence = tokenizer.texts_to_sequences(doc)
sequence
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

sequences = pad_sequences(sequence, padding='post') 
```

```python
sequences
```

```python
vocab_size = len(tokenizer.word_index) + 1
```

```python
from keras.layers import Input
max_len = 5   # example
model = Sequential()
model.add(Input(shape=(max_len,)))   

model.add(Embedding(vocab_size, output_dim=2))
model.add(SimpleRNN(16))    
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(sequences, np.array(labels), epochs=20,validation_split=0.2)
```

#### data is so small

```python
test = ['I love this', 'This is bad']

test_seq = tokenizer.texts_to_sequences(test)
test_pad = pad_sequences(test_seq, maxlen=max_len, padding='post')

pred = model.predict(test_pad)
print(pred)
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,y_train),(X_test,y_test)=imdb.load_data()
```

```python
X_train
```

```python
X_train.shape
```

```python
X_train=pad_sequences(X_train,padding='post')
X_test=pad_sequences(X_test,padding='post')
```

```python
X_train.shape
```

```python
# first review
X_train[0]
```

```python
model = Sequential()

model.add(Embedding(np.max(X_train)+1, 2, input_length=X_train.shape[1]))  

model.add(SimpleRNN(32))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=20,validation_data=(X_test,y_test))
```
